```python
# prompt: buarkan script untuk menghitung luas tabung

def luas_tabung(r, t):
  """
  Menghitung luas tabung.

  Args:
      r: jari-jari tabung (dalam satuan panjang).
      t: tinggi tabung (dalam satuan panjang).

  Returns:
      luas tabung (dalam satuan luas).
  """
  import math
  return 2 * math.pi * r * (r + t)

# Input jari-jari dan tinggi tabung.
r = float(input("25 (m): "))
t = float(input("40 (m): "))

# Hitung luas tabung.
luas = luas_tabung(r, t)

# Print luas tabung.
print("Luas tabung:", luas)
```
